```python
#Se importa la base de datos de pa√≠ses
!pip install geopandas
import os, geopandas as gpd

countries=gpd.read_file(os.path.join("World_Countries"))
```

```python
type(countries)
```

```python
# Se presenta el mapa de todos los pa√≠ses
countries.plot(facecolor="yellow",#este de ac√° es el relleno
               edgecolor='black', #este de ac√° las l√≠neas
               linewidth=0.1) #grosor de las l√≠neasb
```

```python
#Agregando las dem√°s bases de datos...

rivers=gpd.read_file(os.path.join("World_Hydrography"))
cities=gpd.read_file(os.path.join("World_Cities"))
```

```python
#Se grafican las ciudades:
cities.plot(marker='.',
            color='brown',
            markersize=1,
            alpha=0.3) 
```

```python
#Se grafican los rios
rivers.plot(edgecolor='red',
            linewidth=1,
            linestyle='dotted')
```

```python
#Para verificar la proyecci√≥n:
countries.crs==cities.crs==cities.crs
```

```python
#Presentaci√≥n en conjunto:
base = countries.plot(facecolor="white",
                      edgecolor='black',
                      linewidth=0.1,
                      figsize=(12,12))

rivers.plot(edgecolor='red', linewidth=0.4,
            ax=base)
cities.plot(marker='.', color='brown', markersize=1,alpha=0.7,
            ax=base)
```

```python
countries.to_file(os.path.join("worldMap.gpkg"),layer='countryBorders', driver="GPKG")
rivers.to_file(os.path.join("worldMap.gpkg"),layer='riverLines', driver="GPKG")
cities.to_file(os.path.join("worldMap.gpkg"),layer='cityPoints', driver="GPKG")
```

```python
#Presentando el ejemplo de Venezuela:
venezuela=countries[countries.COUNTRY=='Venezuela']
```

```python
citiesVenezuela_clipped = gpd.clip(gdf=cities,
                          mask=venezuela)
riversVenezuela_clipped = gpd.clip(gdf=rivers,
                               mask=venezuela)
```

```python
#Ploteamos
base = venezuela.plot(facecolor="green", edgecolor='black', linewidth=0.4,figsize=(5,5))
citiesVenezuela_clipped.plot(marker='+', color='red', markersize=15,
                    ax=base)
riversVenezuela_clipped.plot(edgecolor='blue', linewidth=0.5,
                    ax=base)
```

```python
#replot
base32619=venezuela.to_crs(32619).plot()
venezuela.to_crs(32619).centroid.plot(color='red',ax=base32619)
```

```python
venezuela_32619=venezuela.to_crs(32619)
cities_venezuela_32619=citiesVenezuela_clipped.to_crs(venezuela_32619.crs)
rivers_venezuela_32619=riversVenezuela_clipped.to_crs(venezuela_32619.crs)
```

```python
import os

venezuela_32619.to_file(os.path.join("venezuelaMaps_32619.gpkg"), layer='country', driver="GPKG")
cities_venezuela_32619.to_file(os.path.join("venezuelaMaps_32619.gpkg"), layer='cities', driver="GPKG")
rivers_venezuela_32619.to_file(os.path.join("venezuelaMaps_32619.gpkg"), layer='rivers', driver="GPKG")
```

```python
#comprobando el centroide:
venezuela_32619.centroid.to_file(os.path.join("venezuelaMaps_32619.gpkg"), layer='centroid', driver="GPKG")
```
